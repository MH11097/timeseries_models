# Model Comparison

Load all model results and compare metrics, predictions, and training times.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.evaluation.comparison import load_results, comparison_table
from src.utils.visualization import plot_metric_comparison

results = load_results('results')
print(f'Found {len(results)} result(s)')
df = comparison_table(results)
display(df)
```

```python
# Metric comparison charts
for metric in ['rmspe', 'rmse', 'mae', 'mape']:
    if metric in df.columns:
        plot_metric_comparison(df, metric=metric)
        plt.show()
```

```python
# Training time comparison
fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot(data=df, x='model_name', y='training_time_seconds', ax=ax)
ax.set_title('Training Time Comparison')
ax.set_ylabel('Seconds')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
